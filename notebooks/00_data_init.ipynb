## DATA DESCRIPTION
Data Contains 37 attributes but some of them are not of the interest for this work. Here are the attributes to be used:
there will be a selection of 22 attributes (2 describing Labeled data Severity) and 2 describing coords.
SEVERITYCODE: 2:Injury / 1:Property Damage        
X           : X-coord location. Not planned to be used     
Y           : Y-coord location. Not planned to be used    
ADDRTYPE    : Intersection / Block      
LOCATION    : NPL location       
SEVERITYDESC: NPL Severity Description       
COLLISIONTYPE: Angle, Sideswipe, Parked Car, etc.       
PERSONCOUNT : # People involved        
PEDCOUNT    : # Pedestrians involved        
PEDCYLCOUNT : # cyclists involved        
VEHCOUNT    : # Vehicles involved        
INCDATE     : Incident Date       
INCDTTM     : Incident Time      
JUNCTIONTYPE: Junction type (Intersection, Midblock, Driveway, etc)       
SDOT_COLCODE: Code of Collision Description
SDOT_COLDESC: NPL Collision Description       
WEATHER     : Weather Condition (Overcast, Rainy, Clear, etc)       
ROADCOND    : Wet / Dry      
LIGHTCOND   : Daylight, Dark, Dark Street Lights on, Dark No street lights, etc)       
SPEEDING    : Y/N       
ST_COLCODE  : Code for description of Collision (Different view)       
ST_COLDESC  : NPL Description       


## DF Drop of non-used Columns

```python
# simply drop whole row with NaN in "price" column
df.drop(['OBJECTID'], axis=1, inplace=True)
df.drop(["INCKEY"], axis=1, inplace=True)
df.drop(["COLDETKEY"], axis=1, inplace=True)
df.drop(["REPORTNO"], axis=1, inplace=True)
df.drop(["STATUS"], axis=1, inplace=True)
df.drop(["INTKEY"], axis=1, inplace=True)
df.drop(["EXCEPTRSNCODE"], axis=1, inplace=True)
df.drop(["EXCEPTRSNDESC"], axis=1, inplace=True)
df.drop(["SEVERITYCODE.1"], axis=1, inplace=True)
df.drop(["INATTENTIONIND"], axis=1, inplace=True)
df.drop(["UNDERINFL"], axis=1, inplace=True)
df.drop(["PEDROWNOTGRNT"], axis=1, inplace=True)
df.drop(["SDOTCOLNUM"], axis=1, inplace=True)
df.drop(["SEGLANEKEY"], axis=1, inplace=True)
df.drop(["CROSSWALKKEY"], axis=1, inplace=True)
df.drop(["HITPARKEDCAR"], axis=1, inplace=True)

# reset index, because we droped two rows
df.reset_index(drop=True, inplace=True)

df.head(5)
```
